```python
import os
import shutil
from glob import glob
import numpy as np
import pandas as pd
import nibabel as nib
from tqdm import tqdm
```

```python
demo = pd.read_csv('/home/coder/data/updated_WSEV/DEMO_250313.csv')
va_path = '/home/coder/data/updated_WSEV/T1_VA'

for idx, row in demo.iterrows():
    subj = row['PTID_ANONY']
    print(subj)
    subj_va = pd.read_csv(os.path.join(va_path, f'{subj}_va.csv'))
    if len(subj_va) != 1:
        raise ValueError(f"{subj} CSV does not have exactly 1 row")
    
    for col in subj_va.columns:
        demo.loc[idx,col] = subj_va.iloc[0][col]

demo.to_csv('/home/coder/data/updated_WSEV/260108_wsev_final_df.csv', index=False)
```
